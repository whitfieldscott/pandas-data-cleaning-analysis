#### Exercise 01. Basic data analysis with native Python structures.  
Create a list with the grades of 5 students: [8.5, 9.2, 7.8, 8.9, 10].

- Calculate the average of the grades.

```python
student_grades = [8.5, 9.2, 7.8, 8.9, 10]
average = sum(student_grades) / len(student_grades)
print(f"The 5 students average grade is {average:.3f}.")
#import pandas as pd
#grade_list = pd.series([8.5, 9.2, 7.8, 8.9, 10])
#grade_list.mean()
```

- Find the highest and lowest grade.

```python
student_grades = [8.5, 9.2, 7.8, 8.9, 10]
highest = max(student_grades)
lowest = min(student_grades)
print(f"The highest grade is {highest} and the lowest grade is {lowest}.")

```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (★★☆)

Print the value obtained from the 'level5' column.

```python
#ds[ds['level5']== 'Arroyomolinos (Madrid)'].head(3)

arroyo_ds = ds[ds.isin(['Arroyomolinos (Madrid)']).any(axis=1)]
avg_arroyo = arroyo_ds['price'].mean()
print(f"The average price of houses in Arroyomolinos (Madrid) is ${avg_arroyo:.2f}.")   

```
